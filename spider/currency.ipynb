## all currency 2012 - 2022

```python
currency = pd.concat([df2012, df2013, df2014, df2015, df2016, df2017, df2018, df2019, df2020, df2021, df2022], axis=0)
currency = currency.reset_index(drop=True)
currency['date'] = pd.to_datetime(currency['date'], format= '%m.%Y')
currency
```

```python
currency.describe()
```

```python
currency.describe().loc[[ 'min',  '25%',  'mean', '50%', '75%', 'max', ]].plot(kind = 'bar', 
                    title = 'Currency Prices 2012 - 2022', figsize=(14,7));
```

```python
currency_melted = currency.melt(id_vars=['date'], var_name='currency', value_name='price')
currency_melted
```

```python
plt.figure(figsize=(14,7))

# pis line on red
plt.axvline(pd.to_datetime('2015-01-01'), color='r', linestyle='--', lw=2)


plt.title('USD & EURO Currency from 2012 to 2022')
sns.lineplot(x = 'date', y = 'price', hue = 'currency', data = currency_melted);
```

```python
currency.to_csv('currency.csv', index=False) 
```
