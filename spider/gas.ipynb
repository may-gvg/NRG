```python
import pandas as pd
import requests
import seaborn as sns
import matplotlib.pyplot as plt
```

## gas for companies
https://pgnig.pl/dla-firmy/taryfa

```python
url = "https://pgnig.pl/dla-firmy/taryfa"
r = requests.get(url)
df = pd.read_html(r.text, thousands=None, decimal=',') 
print(f'Total tables: {len(df)}')
```

Ceny i stawki opłat dla Odbiorców gazu ziemnego wysokometanowego (grupy E) – grupy taryfowe BW

```python
gas_comp = df[0]
gas_comp
```

Ceny i stawki opłat dla Odbiorców gazu ziemnego zaazotowanego (podgrupy Ls) – grupy taryfowe BZ

```python
gas_comp1 = df[1]
gas_comp1
```

Ceny i stawki opłat dla Odbiorców gazu ziemnego zaazotowanego (podgrupy Lw) – grupy taryfowe BS

```python
gas_comp2 = df[2]
gas_comp2
```

```python
gas_comp_all = pd.concat([gas_comp, gas_comp1, gas_comp2], axis=0)
```

```python
gas_comp_all
```

```python
gas_comp_all.columns = ['Grupa taryfowa', 'Paliwo gazowe Netto = Brutto' , 'Paliwo gazowe Brutto',
                        'Paliwo opał NETTO = Brutto',  'Paliwo opał BRUTTO']
```

```python
gas_comp_all = gas_comp_all.reset_index(drop=True)
gas_comp_all = gas_comp_all[['Grupa taryfowa', 'Paliwo gazowe Netto = Brutto', 'Paliwo opał NETTO = Brutto']] 
```

```python
gas_comp_all.at[0,'Grupa taryfowa']= "BW-1,2,3,4"
gas_comp_all.at[4,'Grupa taryfowa']= "BZ-1,2,3,4"
gas_comp_all.at[8,'Grupa taryfowa']= "BS-1,2,3,4"
```

```python
gas_comp_all
```

```python
gas_comp_melted = gas_comp_all.melt(id_vars=['Grupa taryfowa'], var_name='type', value_name='price')
gas_comp_melted = gas_comp_melted.sort_values('price')
```

```python

plt.figure(figsize=(20,10))
plt.ylim([75, 80])
plt.title("Cena paliwa gazowego i opałowego per Grupa taryfowa dla firm")
sns.barplot(data = gas_comp_melted, x = 'type', y = 'price', hue = 'Grupa taryfowa' );
```

## gas for indiviadual users
https://pgnig.pl/dla-domu/taryfa-gazu

```python
url = "https://pgnig.pl/dla-domu/taryfa-gazu"
r = requests.get(url)
df1 = pd.read_html(r.text, thousands=None, decimal=',') 
print(f'Total tables: {len(df1)}')
```

Ceny i stawki opłat dla Odbiorców gazu ziemnego wysokometanowego (grupy E) – grupy taryfowe W

```python
gas_ind = df1[0]
gas_ind
```

Ceny i stawki opłat dla Odbiorców gazu ziemnego zaazotowanego (podgrupy Lw) – grupy taryfowe S

```python
gas_ind1 = df1[1]
gas_ind1
```

Ceny i stawki opłat dla Odbiorców gazu ziemnego zaazotowanego (podgrupy Ls) – grupy taryfowe Z

```python
gas_ind2 = df1[2]
gas_ind2
```

```python
gas_ind_all = pd.concat([gas_ind, gas_ind1, gas_ind2], axis=0)
gas_ind_all.columns = ['Grupa taryfowa', 'Paliwo gazowe Netto' ,
                        'Paliwo opał NETTO',  'Abonament Netto']
gas_ind_all = gas_ind_all.reset_index(drop=True)
```

```python
gas_ind_all
```

```python
abonament = gas_ind_all[['Grupa taryfowa','Abonament Netto']]
plt.figure(figsize=(20,10))

plt.title("Ceny abonamentu per Grupa taryfowa dla indywidualnych")
sns.barplot(data = abonament, x = 'Grupa taryfowa', y = 'Abonament Netto' );
```

```python
abonament = gas_ind_all[['Grupa taryfowa','Abonament Netto']]
plt.figure(figsize=(20,10))
abonament = abonament.sort_values('Abonament Netto')
plt.title("Ceny abonamentu per Grupa taryfowa dla indywidualnych")
sns.barplot(data = abonament, x = 'Grupa taryfowa', y = 'Abonament Netto' );
```

```python
gas_ind_melted = gas_ind_all[['Grupa taryfowa', 'Paliwo gazowe Netto', 'Paliwo opał NETTO']]
```

```python
gas_ind_melted = gas_ind_melted.melt(id_vars=['Grupa taryfowa'], var_name='type', value_name='price')
gas_ind_melted = gas_ind_melted.sort_values('price')
```

```python
plt.figure(figsize=(20,10))
plt.ylim([19.75, 20.55])
plt.title("Cena paliwa gazowego i opałowego per Grupa taryfowa dla ind")
sns.barplot(data = gas_ind_melted, x = 'Grupa taryfowa', y = 'price', hue = 'type' );
```
