## Oil_95

```python
df3 = pd.read_html('https://www.lotos.pl/145/type,oil_95/dla_biznesu/hurtowe_ceny_paliw/archiwum_cen_paliw', 
                   thousands=None, decimal=',')
arch95 = df3[0]

arch95['Data zmiany'] = pd.to_datetime(arch95['Data zmiany'], format= '%Y-%m-%d')

arch95['Cena'] = arch95['Cena'].str.replace(',', '.')
arch95['Cena'] = arch95['Cena'].str.replace(" ", "")
arch95.Cena = arch95.Cena.apply(pd.to_numeric)

arch95['Akcyza'] = arch95['Akcyza'].str.replace(',', '.')
arch95['Akcyza'] = arch95['Akcyza'].str.replace(" ", "")
arch95.Akcyza = arch95.Akcyza.apply(pd.to_numeric)


arch95
```

```python
arch95.describe().plot(kind = 'bar', title = 'PB95 Prices 2004 - 2022', figsize=(14,7));
```

```python
arch95_melted = arch95.melt(id_vars=['Data zmiany'], var_name='Elements', value_name='Value')
arch95_melted
```

```python
plt.figure(figsize=(14,7))

# war red line
plt.axvline(pd.to_datetime('2022-02-17'), color='r', linestyle='--', lw=2)

sns.lineplot(data = arch95_melted, x = 'Data zmiany', y = 'Value', hue = 'Elements' );
```
